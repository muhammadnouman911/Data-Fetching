<h1>Importing Libraries<h1>

Pandas will be used for handeling data.

Requests will be used for fetcing data from api.

```python
import requests
import pandas as pd
```

<h1>Storing the data<h1>

In requests library there is a function called .get() which helps in gathering of data from API.

I'm storing the data in a varible called req.

```python
url=requests.get("https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&per_page=100&page=1")
```

<h1>Converting the web data into JSON<h1>

Converting the web data response into JSON format to ensure it is structured and
 can be easily manipulated or processed by Python for further analysis


```python
url.json()
```

<h1>Creating empty List<h1>

We can create an empty List to store the data collected during the loop for further processing.


```python
dfs =[]
```

```python
dfs
```

<h1>Appling the loop for fetching the data for all pages<h1>

The website contains data across 163 pages.

Using the range function to iterate through these pages.


```python
for i in range(1, 163):
    url = requests.get(f'https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&per_page=100&page={i}')
    temp_df = pd.DataFrame(url.json()) 
    dfs.append(temp_df) 

# Combine all DataFrames in the list into one DataFrame
df = pd.concat(dfs, ignore_index=True)
```

<h1>Exporting it in CSV format<h1>

```python
df.to_csv("cryptocurrency_dataset.csv", index=False)

print("Dataset saved as 'cryptocurrency_dataset.csv'")
```
